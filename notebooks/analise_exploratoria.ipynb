# Análise do valor do aluguel por m² em São Paulo

Este projeto analisa dados de imóveis para aluguel em São Paulo,
com foco no valor médio do aluguel por metro quadrado, considerando
bairro e tipo de imóvel.


## Objetivo

- Calcular o valor do aluguel por m²
- Analisar diferenças entre tipos de imóveis
- Comparar tipos de imóveis dentro do mesmo bairro
- Identificar padrões globais entre bairros


```python
import sys
sys.path.append('../src')

from data_loader import preparar_dados
from analyses import (
    analise_tipo_imovel,
    analise_bairro_tipo,
    analise_bairro_tipo_matriz
)
from visualizations import (
    plot_tipo_imovel,
    plot_bairro_tipo,
    plot_heatmap
)
```

## Preparação dos dados

Nesta etapa, os dados são carregados, filtrados para remover bairros
com baixa representatividade e é criada a métrica de valor por m².


```python
dados = preparar_dados('../data/data.csv')
dados.head()
```

## Análise 1 — Valor do aluguel por m² por tipo de imóvel

Esta análise apresenta uma visão geral do mercado imobiliário,
comparando o valor médio do aluguel por m² entre diferentes tipos de imóveis.


```python
resultado_tipo = analise_tipo_imovel(dados)
plot_tipo_imovel(resultado_tipo)
```

## Análise 2 — Comparação por tipo dentro de um bairro

Para evitar o viés de localização, esta análise compara os tipos de imóveis
dentro de um mesmo bairro.


```python
bairro = 'Vila Mariana'
resultado_bairro = analise_bairro_tipo(dados, bairro)
plot_bairro_tipo(resultado_bairro, bairro)
```

## Análise 3 — Padrão global por bairro e tipo

O heatmap permite visualizar padrões do valor do aluguel por m²
considerando simultaneamente bairro e tipo de imóvel.


```python
tabela = analise_bairro_tipo_matriz(dados)
plot_heatmap(tabela)
```

## Conclusões

- Studios e kitnets apresentam os maiores valores de aluguel por m²
- Casas possuem menor custo relativo por área
- Mesmo dentro do mesmo bairro, o tipo do imóvel influencia fortemente o preço
- O padrão se repete em diversos bairros com alta oferta de imóveis
